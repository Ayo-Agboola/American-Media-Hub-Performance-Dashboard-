```python
import pandas as pd
df = pd.read_csv("SLU Opportunity Wise Data.csv")
```

Age Column

```python
df["age_of_learner"] = (
    (pd.Timestamp("today") - df["date_of_birth"])
    .dt.days // 365

)
```

Validate

```python
df["age_of_learner"].describe()
```

```python
df["age_of_learner"].isna().sum()
```

Engagement Duration

```python
df["engagement_duration_days"] = (df["opportunity_start_date"] -
df["apply_date"]).dt.days
```

```python
df['engagement_duration_days'].describe()
```

```python
df["engagement_duration_days"].isna().sum()
```

Flag column

```python
df["negative_engagement_flag"] = df["engagement_duration_days"] < 0
```

```python
df["negative_engagement_flag"].value_counts()
```
